```python
import seaborn as sns
import numpy as np
# import pandas as pd

sns.get_dataset_names()
```

```python
penguins_data = sns.load_dataset("penguins")

penguins_data.info()
penguins_data.head()
```

```python
sns.set_style("ticks") # ticks, white, whitegrid, dark, darkgrid
sns.set_context("notebook")

sns.scatterplot(data= penguins_data, x= "flipper_length_mm" ,y = "body_mass_g", hue="island", style="sex"
                # , palette="Set1"  , alpha=0.5
                )
sns.despine(left=True)
```

```python
sns.stripplot(data= penguins_data, x= "flipper_length_mm" ,y = "body_mass_g", hue="island" , dodge=True, jitter=True)
```

```python
sns.set_context("talk")
sns.swarmplot(data= penguins_data, x= "flipper_length_mm" ,y = "body_mass_g", hue="island" )
sns.despine()
```

```python
sns.set_context("notebook")
# sns.histplot(data= penguins_data, x= "body_mass_g")
sns.histplot(data= penguins_data, x= "body_mass_g", hue="sex", multiple="stack")
```

```python
sns.regplot(data=penguins_data, x= "flipper_length_mm" ,y = "body_mass_g", color="green", scatter=False)
```

```python
sns.barplot(data=penguins_data, x= "island" ,y = "body_mass_g", hue="sex", palette=["green", "orange"], estimator=np.sum)
```

```python
sns.set_style("whitegrid")
sns.lineplot(data=penguins_data, x= "flipper_length_mm" ,y = "body_mass_g", color="green", hue ="sex")
```

```python
sns.jointplot(data=penguins_data, x= "flipper_length_mm" ,y = "body_mass_g", hue="sex", palette=["green","orange"],
            #   kind="kde"
              )
```

```python
sns.countplot(data=penguins_data, x= "island" , hue="species", palette="Set2")
```

```python
sns.boxplot(data=penguins_data, x="species" , y="body_mass_g", hue= "sex")
```

```python
sns.violinplot(data=penguins_data, x="species", y="body_mass_g")
sns.swarmplot(data=penguins_data, x="species", y="body_mass_g", hue="island" , size=3)
```

```python
sns.kdeplot(data=penguins_data, x="body_mass_g", hue= "island", fill=True)
```

```python
sns.heatmap(data= penguins_data[["bill_length_mm","bill_depth_mm","flipper_length_mm"]].corr(), annot=True, cmap="Blues" )
```

```python
sns.rugplot(data=penguins_data, y= "body_mass_g", hue = "species")
```

```python
sns.pairplot(data=penguins_data, hue= "sex")
```
